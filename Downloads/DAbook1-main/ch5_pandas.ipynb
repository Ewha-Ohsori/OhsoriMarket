```python
import pandas as pd
df = pd.read_csv('materials/5장/vaccine.csv',encoding = 'cp949')
df
```

```python

df.tail()
```

```python

df = df.drop(['date', 'day1sum', 'day1rate', 'day2sum', 'day2rate'], axis=1)
df.head()
```

```python
count = df.count()
mean = df.mean()
std = df.std()
min = df.min()
max = df.max()
df_25 = df/4

summary_df = pd.DataFrame({
    'count': count,
    'mean': mean,
    'std': std,
    'min': min,
    #'25%': df_25,
    'max': max
})

summary_df = summary_df.T

print(summary_df)
```

```python

summary = df.describe()
summary = summary.T
print(summary_df)
```
