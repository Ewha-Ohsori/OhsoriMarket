# series 만들기

```python
# 5-1 정수형 series 만들기

import pandas as pd
score = pd.Series([95,90,85,90,95]) # 정수 인덱스 자동 생성
score
```

```python
# 5-2 실수형 series 만들기
s=pd.Series([4,3.5,3.8,3,3.7])
s
```

```python
s=pd.Series([90,80,95],index=['A','B','C'])
s
```

# Dataframe 만들기

```python
# 5-4 넘파이로 배열 생성
import numpy as np
import pandas as pd
arr = np.arange(12).reshape(4,3)
print(arr)
print()
df=pd.DataFrame(arr)
print(df)
```

```python
# 5-5 딕셔너리로 배열 생성
df = pd.DataFrame({'name' : ['장은실', '오경선', '양숙희'],
                   'score' : [90,80,95],
                   'dept':['com','eng','math']})    # dataframe의 열(column)은 서로 다른 자료형 (dtype)을 갖는다.
df
```
